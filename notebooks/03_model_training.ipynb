```python
df = df.fillna(df.median(numeric_only=True))
```

```python
# SORT DATA (VERY IMPORTANT)
df = df.sort_values(by=['Abbreviation', 'Year', 'Round'])

# ---- BASIC ROLLING FEATURES ----

df['AvgFinish_Last3'] = (
    df.groupby('Abbreviation')['Position']
    .shift(1)
    .rolling(3)
    .mean()
)

df['AvgFinish_Last5'] = (
    df.groupby('Abbreviation')['Position']
    .shift(1)
    .rolling(5)
    .mean()
)

df['DNFRate_Last5'] = (
    df.groupby('Abbreviation')['DNF']
    .shift(1)
    .rolling(5)
    .mean()
)

# ---- MOMENTUM ----

df['PositionChange'] = df['GridPosition'] - df['Position']

df['Momentum_Last3'] = (
    df.groupby('Abbreviation')['PositionChange']
    .shift(1)
    .rolling(3)
    .mean()
)

# ---- GRID NON-LINEAR ----

df['GridSquared'] = df['GridPosition'] ** 2

# ---- TEAM FORM ----

df['TeamForm'] = (
    df.groupby('TeamName')['Position']
    .shift(1)
    .rolling(3)
    .mean()
)

# ---- CONSISTENCY ----

df['Consistency'] = (
    df.groupby('Abbreviation')['Position']
    .shift(1)
    .rolling(5)
    .std()
)
```
